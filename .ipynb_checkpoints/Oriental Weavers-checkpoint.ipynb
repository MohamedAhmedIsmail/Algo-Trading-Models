```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import preprocessing,cross_validation,svm
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
oriental_Weavers=pd.read_csv("C:/Users/mohamed ismail/Desktop/GP And Data/Oriental Weavers.csv",index_col=1)
```

```python
oriental_Weavers.head()
```

```python
oriental_Weavers=oriental_Weavers[-500:]
oriental_Weavers.head()
```

```python
oriental_Weavers['EWMA_12']=oriental_Weavers['CLOSE_PRICE'].ewm(span=12).mean()
oriental_Weavers['HL_PCT']=(oriental_Weavers['HIGH_PRICE']-oriental_Weavers['LOW_PRICE'])/oriental_Weavers['LOW_PRICE']*100.0
oriental_Weavers['PCT_change']=(oriental_Weavers['CLOSE_PRICE']-oriental_Weavers['OPEN_PRICE'])/oriental_Weavers['OPEN_PRICE']*100.0
```

```python
oriental_Weavers.head()
```

```python
tmp=oriental_Weavers[['EWMA_12','HL_PCT','PCT_change']].plot(figsize=(18,6))
plt.show()
```

```python
oriental_Weavers.dropna(inplace=True)
len(oriental_Weavers)
```

```python
oriental_Weavers['CLOSE_PRICE'].shift(-5).tail()
```

```python
X=oriental_Weavers[['EWMA_12','HL_PCT','PCT_change']].values
X
```

```python
y=oriental_Weavers['CLOSE_PRICE'].values
y
```

```python
len(X),len(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,shuffle=False,random_state=1)
```

```python
len(X),len(X_train),len(X_test)
```

```python
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
LR=LinearRegression()
KNN=KNeighborsRegressor()
SVR=svm.SVR()
```

```python
LR.fit(X_train,y_train)
KNN.fit(X_train,y_train)
SVR.fit(X_train,y_train)
```

```python
y_pred_LR=LR.predict(X_test)
y_pred_KNN=KNN.predict(X_test)
y_pred_SVR=SVR.predict(X_test)
len(y_pred_LR)
```

```python
oriental_Weavers['LR_pred']=np.nan
oriental_Weavers['KNN_pred']=np.nan
oriental_Weavers['SVR_pred']=np.nan
oriental_Weavers['LR_pred'].iloc[350:]=y_pred_LR
oriental_Weavers['KNN_pred'].iloc[350:]=y_pred_KNN
oriental_Weavers['SVR_pred'].iloc[350:]=y_pred_SVR
```

```python
oriental_Weavers.tail()
```

```python
oriental_Weavers[['CLOSE_PRICE','LR_pred','KNN_pred','SVR_pred']].plot(figsize=(18,6))
plt.show()
```

```python
oriental_Weavers[['CLOSE_PRICE','LR_pred','KNN_pred','SVR_pred']][400:].plot(figsize=(18,6))
plt.show()
```

```python
LR_RMSE=np.sqrt(mean_squared_error(y_true=y_test,y_pred=y_pred_LR))
KNN_RMSE=np.sqrt(mean_squared_error(y_true=y_test,y_pred=y_pred_KNN))
SVR_RMSE=np.sqrt(mean_squared_error(y_true=y_test,y_pred=y_pred_SVR))
```

```python
print('Linear Regression :',LR_RMSE)
print('KNN :',KNN_RMSE)
print('SVR :',SVR_RMSE)
```

```python
accuracy_LR=LR.score(X_test,y_test)
accuracy_LR*100.0
```

```python
accuracy_KNN=KNN.score(X_test,y_test)
accuracy_KNN*100.0
```

```python
accuracy_SVR=SVR.score(X_test,y_test)
accuracy_SVR*100.0
```
